## Data Preparation

In preparation for this analysis, I plan on gathering as much info about each column and value in the dataset. If there are columns that aren't related to my analysis, or columns with too much missing or null data, I will remove the column entirely. For columns that could assist in my analysis, I will determine what imputation methods will help create an accurate visualization. The method I have decided to go with first is to gather a percentage of missing data for each column. Anything with 50% or more missing data will be considered for column removal. If a column with 50% or more missing data will benefit the analysis, I will consider alternative methods to make sure it can be used. 

After investigating further, I have determined that the data remaining after dropping columns that are unrelated to this analysis will be ideal for continuing in the process of modeling this data. 

```python
# Checking for missing values in the dataset.
missing_values = aviation_data.isna().sum()
print(missing_values)
```

```python
# Calculating the percentage of missing values in each column of the dataset.
missing_percent = aviation_data.isnull().sum() / len(aviation_data) * 100
print(missing_percent)
```

```python
# I want to make sure each name is unique so I am checking the value counts of the names column. I will eliminate spaces and make all the names uppercase. 
aviation_data['Model'] = aviation_data['Model'].str.strip().str.upper()
```

```python
# I want to determine if I can fill the 'Aircraft_Category' columnn with information I can gather from the 'Model' column. So I will print the value count of the 'Model' to see if I can use it as a cross reference.
model_counts = aviation_data['Model'].value_counts()
print(model_counts)
```
